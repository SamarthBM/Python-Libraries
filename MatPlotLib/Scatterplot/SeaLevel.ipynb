```python
import matplotlib.pyplot as plt
import pandas as pd
from LogHandler import logger

try:

    data = pd.read_csv('sealevel.csv')

    year = data['year']
    sea_level = data['sea_level']

    plt.scatter(year, sea_level, edgecolors='g')

    plt.xlabel('Year' ,fontsize = 15)  
    plt.ylabel('Sea Level (inches)',fontsize = 15)
    plt.title('Rise in Sea level',fontsize = 15)
    
    plt.show()

except Exception as e:
    logger.error(e)
```
