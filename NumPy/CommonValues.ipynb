```python
import numpy as np
from LogHandler import logger

try:
    
    arr1 = np.array([18,17,15,9,10])
    arr2 = np.array([15,25,11,18])

    print("Common values present in both arrays are:", np.intersect1d(arr1,arr2))

except Exception as e:
    logger.error(e)
```
