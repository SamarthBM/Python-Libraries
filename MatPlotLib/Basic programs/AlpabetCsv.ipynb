```python
import matplotlib.pyplot as plt
from LogHandler import logger
import pandas as pd
  
try:
    
    data = pd.read_csv("AlphabetInc.csv", sep=',', parse_dates=True, index_col=0)
    data.plot(color="blue", linewidth = 5, linestyle = 'dashed')

    plt.xlabel('Dates')
    plt.ylabel('Close')
    plt.title('Alphabet Inc')
    
    plt.legend()
    plt.grid()
    plt.show()

except Exception as e:
    logger.error(e)
```
