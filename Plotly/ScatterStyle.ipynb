```python
import pandas as pd
import plotly.express as px
from pylab import randn
from LogHandler import logger

try:

    x = randn(500)
    y = randn(500)
    df = pd.DataFrame({ 'x-axis':x, 'y-axis':y })

    fig = px.scatter(df, x='x-axis', y='y-axis', width=700, height=500, title="Scatter Plot",color="y-axis")
    fig.show()

except Exception as e:
    logger.error(e)
```
