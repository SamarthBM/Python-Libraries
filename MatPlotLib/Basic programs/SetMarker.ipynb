```python
from matplotlib import pyplot as plt
from LogHandler import logger

try:
    
    x = (1,2,5,4,8,6)
    y = [num**2 for num in x]
    print(x)
    print(y)

    plt.plot(x, color = "green", linewidth = 3,label = "Square of numbers", linestyle = "dotted", marker = 'o',
             markerfacecolor = "blue")

    plt.plot(y, color = "red", linewidth = 3 ,label = "Square of numbers",  linestyle = "dashdot",marker = 'o',
             markerfacecolor = "blue")

    plt.xlabel("X-axis")
    plt.ylabel("y-axis")
    plt.title("Two lines on same plot with legend")

    plt.xlim(0, 7)  # Setting limit for x axis.
    plt.ylim(0, 70) # Setting limit for y axis.
    
    plt.legend()
    plt.show()

except Exception as e:
    logger.error(e)
```
