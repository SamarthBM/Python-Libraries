```python
import matplotlib.pyplot as plt
import pandas as pd
from LogHandler import logger

try:

    df =  pd.read_csv('CountryMedals.csv')
    country_data = df["country"]
    medal_data = df["gold_medal"]

    explode = (0.2, 0, 0, 0, 0)  

    plt.pie(medal_data, labels=country_data, explode=explode,startangle=90)

    plt.title("Country with their total gold medals")
    plt.show()

except Exception as e:
    logger.error(e)
```
