```python
from matplotlib import pyplot as plt
from LogHandler import logger

try:
    
    x = (1,2,3,4,5,6)
    y = [num**2 for num in x]
    print(x)
    print(y)

    plt.plot(x, color = "green", linewidth = 4 ,label = "Numbers")
    plt.plot(y, color = "red", linewidth = 7 ,label = "Square of numbers")

    plt.xlabel("X-axis")
    plt.ylabel("y-axis")
    plt.title("Two lines on same plot with legend")

    plt.legend()
    plt.show()

except Exception as e:
    logger.error(e)
```
