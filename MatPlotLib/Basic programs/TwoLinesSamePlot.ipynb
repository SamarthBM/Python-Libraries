```python
from matplotlib import pyplot as plt
from LogHandler import logger

try:
    x = (1,2,3,4,5,6)
    # y = (1,4,9,16,25,36)
    y = [num**2 for num in x]

    print(x)
    print(y)

    plt.plot(x, label = "Numbers")
    plt.plot(y, label = "Square of number")

    plt.xlabel("X-axis")
    plt.ylabel("y-axis")
    plt.title("Two lines on same plot with legend")

    plt.legend()
    plt.show()

except Exception as e:
    logger.error(e)
```
